Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Answer : Lookup Error is the base class for key error and index error. The key error is raised when a wrong key is used to access a dictionary value, using a key that is not in the dictionary.This error means that the key you use to access a dictionary is wrong.

The index error is raised when you try to access a sequence (index) of a list that does not exist or is out of range.

```python
#key error
try:
    d = {'a':[1,2,3,4],'key':'sudh'}
    d['key10']
except KeyError as e :
    print (e)
```

```python
#index error
try:
    l = [1,2,3,4,5]
    l[10]
except IndexError as e:
    print(e)
```
